```python
import numpy as np
from IPython.display import Image
from sklearn.metrics import confusion_matrix

from sklearn.metrics import (
precision_score, 
recall_score,
f1_score
)
```

```python
y_true = [0, 0, 0, 1, 1, 1]
y_pred = [1, 0, 0, 1, 1, 1]
```

```python
confmat = confusion_matrix(y_true, y_pred)
```

```python
confmat

# TN  FP
# FN  TP
```

```python
tp = confmat[1][1]
fp = confmat[0][1]
fn = confmat[1][0]
tn = confmat[0][0]

precision = tp / (tp + fp)
recall = tp / (tp + fn)

acc = (tp + tn) / (tp + fp + tn + fn)
```

```python
print('precision :', precision)
print('recall :', recall)
print('acc :', acc)
```

```python
y_true = [0, 0, 0, 1, 1, 1]
y_pred = [1, 0, 0, 1, 1, 1]
```

```python
precision_sklearn = precision_score(y_true, y_pred)
recall_sklearn = recall_score(y_true, y_pred)

print('precision :', precision_sklearn)
print('recall :', recall_sklearn)
```
