```python
from sklearn.datasets import make_classification
```

- 임의로 데이터 셋과 피쳐 수를 구한다.

```python
x, y = make_classification(n_samples = 500, n_features = 5, n_redundant = 0, random_state = 0)
```

```python
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np
%matplotlib inline
```

```python
x.mean(), y.mean()
```

```python
fig, axs = plt.subplots(ncols = 1, figsize = (10, 10), 
                       subplot_kw = {'projection': '3d'})

x_data = np.linspace(-2.35, 2.35, 100)
y_data = np.linspace(-2.35, 2.35, 100)
X, Y = np.meshgrid(x_data, y_data)
Z = X + Y

axs.plot_surface(X, Y, Z, cmap = 'bone',alpha = 1)
axs.scatter(x[:, 2], x[:, 1], x[:, 4], c = y);
```
